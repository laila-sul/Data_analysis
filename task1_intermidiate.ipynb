```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
data1 = pd.read_csv('globalterrorismdb_0718dist.csv',low_memory=False)
```

step2: Data understanding in our case the dataset has various types of structured data

```python
data1.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1.columns
```

```python
data1.describe()
```

```python
data = data1[['iyear','imonth','iday','country_txt','provstate','region_txt','city','latitude','longitude',"attacktype1_txt",'nkill',
               'nwound','target1','summary','gname','targtype1_txt','weaptype1_txt','motive']]
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data['nkill'].fillna(data['nkill'].mode()[0])
```

```python
data['nwound'].fillna(data['nwound'].mode()[0])
```

```python
data["provstate"].fillna( method ='ffill', inplace = True)
```

```python
data["city"].fillna("No city", inplace = True) 
```

```python
data['latitude'].fillna(data['latitude'].mode()[0])
```

```python
data['longitude'].fillna(data['longitude'].mode()[0])
```

```python
data["target1"].fillna("No target", inplace = True) 
```

```python
data["summary"].fillna("No summary", inplace = True) 
```

```python
data["motive"].fillna("No motive", inplace = True) 
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
Casualities = data.nkill + data.nwound
data.Casualities=Casualities
```

```python
data.head()
```

```python
print(data["Casualities"])
```

```python
data.describe()
```

```python
year = data['iyear'].unique()
years_count = data['iyear'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (16,8))
sns.barplot(x = year,
           y = years_count,
           palette = "tab10")
plt.xticks(rotation = 50)
plt.xlabel('Attacking Year',fontsize=20)
plt.ylabel('Number of Attacks Each Year',fontsize=20)
plt.title('Attacks In Years',fontsize=30)
plt.show()
```

```python
pd.crosstab(data.iyear, data.region_txt).plot(kind='area',stacked=False,figsize=(20,10))
plt.title('Terrorist Activities By region In Each year',fontsize=25)
plt.ylabel('Number of Attacks',fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.show()
```

```python
data[['iyear','Casualities']].groupby(["iyear"],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['cyan'])
plt.xticks(rotation=50)
plt.title("Number of Casualities per year  ",fontsize=20)
plt.ylabel('year',fontsize=15)
plt.xlabel('Casualities',fontsize=15)
plt.show()
```

```python
data[['attacktype1_txt','Casualities']].groupby(["attacktype1_txt"],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['cyan'])
plt.xticks(rotation=50)
plt.title("Number of Casualities per attack type  ",fontsize=20)
plt.ylabel('Casualities',fontsize=15)
plt.xlabel('Attack Type',fontsize=15)
plt.show()
```

```python
plt.subplots(figsize=(20,10))
x=data.attacktype1_txt
sns.countplot(x=x,order=x.value_counts().index);
plt.xlabel("Attacktype",fontsize=15)
plt.ylabel("count",fontsize=15)
plt.title("Attack per year",fontsize=20)
plt.show()
```

```python
data['country_txt'].value_counts().to_frame().sort_values('country_txt',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10),color='blue')
plt.xticks(rotation = 50)
plt.xlabel("City",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Top 10 most effected city",fontsize=20)
plt.show()
```

```python
data['gname'].value_counts().to_frame().drop('Unknown').sort_values('gname',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10),color='blue')
plt.xticks(rotation = 50)
plt.xlabel("Attacker group name",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Top 10 most Actif attack group",fontsize=20)
plt.show()
```

```python
data[['gname','nkill']].groupby(['gname'],axis=0).sum().drop('Unknown').sort_values('nkill',ascending=False).head(10).plot(kind='bar',color='blue',figsize=(20,10))
plt.xticks(rotation = 50)
plt.xlabel("Attacker group name",fontsize=15)
plt.ylabel("Number of killed people",fontsize=15)
plt.title("Top 10 most Actif attack group",fontsize=20)
plt.show()
```

```python
data['target1'].value_counts().to_frame().drop('Unknown').sort_values('target1',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10),color='yellow')
plt.xticks(rotation = 50)
plt.xlabel("Attacked target",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Top 10 most attacked target",fontsize=20)
plt.show()
```

```python
df=data[['gname','country_txt','nkill']]
df=df.groupby(['gname','country_txt'],axis=0).sum().sort_values('nkill',ascending=False).drop('Unknown').reset_index().head(10)
df
```
